```python
#Bibliotecas
import pandas as pd
import numpy as np
from faker import Faker
import random
```

```python
#Produto
df= pd.DataFrame()
faker = Faker()

# Gerar 50 linhas com informações de produtos fictícios
for _ in range(200):
    preco = round(faker.pydecimal(left_digits=3, right_digits=2, positive=True), 2)
    df = df.append({
        'IdProduto': faker.ean13(),
        'DescProduto': faker.word().capitalize(),
        'Quantidade': faker.random_int(min=1, max=1000),
        'Preço': faker.random_number(digits=2) + faker.random_number(digits=1) / 10 + faker.random_number(digits=1) / 100
    }, ignore_index=True)

df
```

```python
faker = Faker()
# Criando e atualizando a df2 - Marca
df2 = pd.DataFrame(columns=['IdMarca','DescMarca','IdProduto','DescProduto', 'Quantidade', 'Preço'])

for _ in range(10):
  df2 = df2.append({
        'IdMarca': random.randint(1, 10),
        'DescMarca': faker.company(),
        'IdProduto': random.randint(1, 10),
        'DescProduto': faker.word().capitalize(),
        'Quantidade': random.randint(1, 500),
        'Preço': faker.random_number(digits=2) + faker.random_number(digits=1) / 10 + faker.random_number(digits=1) / 100
    }, ignore_index=True)

#Cliente
df3 = pd.DataFrame(columns=['IdCliente','Nome','Idade','Email', 'País', 'Cidade', 'IdProduto', 'Quantidade', 'IdMarca'])

for _ in range(10000):
  df3 = df3.append({
    'IdCliente': random.randint(10000000, 99999999),
    'Nome': faker.name(),
    'Idade': random.randint(1, 60),
    'Email': faker.email(),
    'País': faker.country(),
    'Cidade': faker.city(),
    'IdProduto': random.randint(1, 10),
    'Quantidade': random.randint(1, 10),
    'IdMarca': random.randint(1, 10)
}, ignore_index=True)


```
